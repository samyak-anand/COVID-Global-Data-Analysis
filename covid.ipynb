```python
synt_col= ['Cough_symptoms', 'Fever', 'Sore_throat',
       'Shortness_of_breath', 'Headache']
```

```python
for col in synt_col :
    cp_data[col].fillna(method='ffill',inplace=True)
```

```python
cp_data.isnull().sum()
```

```python
# Fill missing 'Sex' values: first forward fill, then set any remaining to 'male'
cp_data['Sex'].fillna(method='ffill', inplace=True)
cp_data['Sex'].fillna('male', inplace=True)
```

```python
# Fill missing 'Age_60_above': first forward fill, then set any remaining to 'Yes'
cp_data['Age_60_above'].fillna(method='ffill', inplace=True)
cp_data['Age_60_above'].fillna('Yes', inplace=True)
```

```python
# Final missing values check
print("\nMissing after cleaning:\n", cp_data.isnull().sum())
print("\nFinal shape:", cp_data.shape)
```
